```r
# Define the list of required R packages
r_packages <- c(
  "DESeq2", "edgeR", "limma", "Glimma", "GEOquery", 
  "Seurat", "tidyverse", "clusterProfiler", "AnnotationDbi", 
  "ggplot2", "UpSetR", "RColorBrewer", "gplots"
)

# Function to check if packages are installed and loadable
check_r_packages <- function(pkg_list) {
  results <- sapply(pkg_list, function(p) {
    if (!requireNamespace(p, quietly = TRUE)) {
      return("MISSING")
    } else {
      library(p, character.only = TRUE)
      return("OK")
    }
  })
  return(results)
}

# Run the check
cat("Checking R Packages:\n")
check_r_packages(r_packages)

# Special check for the Arabidopsis annotation package (since you switched to plants)
if (!requireNamespace("org.At.tair.db", quietly = TRUE)) {
  cat("\nNote: 'org.At.tair.db' is missing. Install it for plant analysis using:\n")
  cat("BiocManager::install('org.At.tair.db')\n")
}
```
